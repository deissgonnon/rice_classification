```python
import pandas as pd
import tensorflow as tf
import os
from tensorflow.keras.preprocessing.image import ImageDataGenerator 
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import ImageGrid
from pathlib import Path
from PIL import Image
import cv2
from tensorflow.keras import layers
from sklearn import preprocessing
```

**Data loading and name encoding**

```python
from pathlib import Path
from sklearn import preprocessing
import os

# Get the current working directory
current_directory = Path(os.getcwd())

# Define the path to the rice_img folder
dataset_path = current_directory / "rice_img"

# Get all image paths from the rice_img folder
image_link = list(dataset_path.glob('**/*.jpg'))

# Extract folder names as image names
image_name = [x.parents[0].stem for x in image_link]

# Encode the image labels
image_label = preprocessing.LabelEncoder().fit_transform(image_name)

# Print the results for verification
print(f"Number of images found: {len(image_link)}")
print(f"Image labels: {image_label}")
```

**df(DataFrame)**

```python
import numpy as np
df = pd.DataFrame()
df['link'] = np.array(image_link, dtype=str)  # Use built-in str type
df['name'] = image_name
df['label'] = image_label

# Print the DataFrame for verification
print(df.head())
```

**Plot the varieties**

```python
df.name.value_counts().plot(kind = 'bar', figsize = (12, 8), grid = True, color = 'blue')
```

**The different varieties are showing bellow**

```python
fig = plt.figure(1, figsize = (15,15))
grid = ImageGrid(fig, 121, nrows_ncols = (5,4), axes_pad = 0.10)
i = 0
for category_id, category in enumerate(df.name.unique()):
    for filepath in df[df['name'] == category]['link'].values[:4]:
        ax = grid[i]
        img = Image.open(filepath)
        ax.imshow(img)
        ax.axis('off')
        if i % 4 == 4-1:
            ax.text(300,100, category, verticalalignment = 'center', fontsize = 20, color ='black')
        i+=1
        
plt.show()
```

```python
train_df, test_df = train_test_split(df, test_size=0.3, random_state = 1) 
```

```python
# Create training and test image generators from the DataFrame. 
# - Each row of the DataFrame provides the file path ('link') and label ('name').
# - 'color_mode' specifies how images are loaded (RGB here).
# - 'target_size' resizes images to 28x28.
# - 'class_mode' defines how labels are handled (categorical, for multi-class classification).
# - 'batch_siz' specifies how many samples per batch.
train_images = ImageDataGenerator().flow_from_dataframe(
    dataframe=train_df,
    x_col='link',
    y_col='name',
    color_mode='rgb',
    batch_siz=32,
    target_size=(28, 28),
    clas_mode='categorical',
    subset='training'
)

test_images = ImageDataGenerator().flow_from_dataframe(
    dataframe=test_df,
    x_col='link',
    y_col='name',
    color_mode='rgb',
    batch_siz=32,
    target_size=(28, 28),
    clas_mode='categorical'
)
```

```python
model = tf.keras.models.Sequential([
    # 1. Define the input layer, specifying the shape (28x28, 3 channels).
    tf.keras.layers.Input(shape=(28, 28, 3)),

    # 2. First convolutional layer with 16 filters, kernel size of 3x3, ReLU activation.
    tf.keras.layers.Conv2D(16, (3, 3), activation='relu'),

    # 3. Downsampling using a 2x2 Max Pool, reducing spatial dimensions.
    tf.keras.layers.MaxPool2D(pool_size=(2, 2)),

    # 4. Second convolutional layer with 32 filters, kernel size of 3x3, ReLU activation.
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu'),

    # 5. Another 2x2 Max Pool for further downsampling.
    tf.keras.layers.MaxPool2D(pool_size=(2, 2)),

    # 6. Flatten the feature maps into a 1D vector for the fully connected layers.
    tf.keras.layers.Flatten(),

    # 7. Dense (fully connected) layer with 64 units and ReLU activation.
    tf.keras.layers.Dense(64, activation='relu'),

    # 8. Dense layer with 128 units and ReLU activation.
    tf.keras.layers.Dense(128, activation='relu'),

    # 9. Output layer with 5 units (classes) and softmax activation for multi-class classification.
    tf.keras.layers.Dense(5, activation='softmax')
])

```

```python
from tensorflow.keras.utils import plot_model

plot_model(
    model,
    to_file='model_plot.png',   # The output file name
    show_shapes=True,           # Display input/output shapes
    show_layer_names=True       # Display layer names
)
```

```python
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# 1. Model compilation:
#    - Using Adam optimizer with a specified learning rate for potentially smoother convergence.
#    - Loss set to 'categorical_crossentropy' for multi-class classification.
#    - Metrics include 'accuracy' to track how well the model predicts.
model.compile(
    optimizer=Adam(learning_rate=1e-3),  # You can tune this learning rate.
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# 2. callback list:
#    - EarlyStopping: stop training early if validation loss stops improving (prevents overfitting).
#    - ReduceLROnPlateau: reduce the learning rate if validation loss plateaus, giving the model
#      another chance to converge.
callbacks_list = [
    EarlyStopping(
        monitor='val_loss',     # What metric to monitor
        patience=3,             # How many epochs to wait for improvement before stopping
        restore_best_weights=True
    ),
    ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.2,            # How much to reduce the LR by
        patience=2             # How many epochs with no improvement before reducing
    )
]

# 3. Fit the model:
#    - Training data: train_images (flow_from_dataframe generator in your example).
#    - Set epochs to 10 (or more, if you like). 3 might be too few for many problems, but it depends on your dataset.
#    - (Optional) Provide validation_data for validation metrics and to enable EarlyStopping/ReduceLROnPlateau.
history = model.fit(
    train_images,
    epochs=10,                # Increase from 3 to 10 for more training (adjust as needed).
    validation_data=test_images,  # Provide a validation set to make callbacks meaningful.
    callbacks=callbacks_list
)
```

```python
# Evaluate the trained model on the test dataset (test_images) to measure loss and accuracy.
model.evaluate(test_images)
```
